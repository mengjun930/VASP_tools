```python
def read_fix_list(filename, n):
    read_fix_list = open(filename, 'r')
    xyz = read_fix_list.readlines()
    read_fix_list.close()
    xyz = xyz[8:]
    read_z = [float(i.split()[2])  for i in xyz]
    fix_list = np.unique(read_z)
    return fix_list[:n]
```

```python
read_fix_list('POSCAR.vasp.bk', 2)
```
